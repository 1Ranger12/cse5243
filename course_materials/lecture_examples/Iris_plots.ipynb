## Lets Look at Iris. Again.

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
import scipy.stats as stat

iris = load_iris()
```

```python
df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
df['target'] = pd.Series(iris['target'], name = 'target_values')
df['target_name'] = df['target'].replace([0,1,2],[species for species in iris['target_names'].tolist()])
df
```

```python
df.columns
```

```python
j=sns.jointplot(x='sepal length (cm)',y='sepal width (cm)',data=df,height=5)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
sns.FacetGrid(df,hue='target_name',size=5)\
.map(plt.scatter,'sepal length (cm)','sepal width (cm)')\
.add_legend()
```

```python
ax=sns.boxplot(x='target_name',y='sepal length (cm)',data=df)
ax=sns.stripplot(x='target_name',y='sepal length (cm)',data=df,jitter=True,edgecolor='gray')
```

```python
sns.violinplot(x='target_name',y='sepal length (cm)',data=df,size=6)
```

```python
sns.pairplot(data=df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)','target_name']], hue = 'target_name', kind='scatter')
```

```python
plt.figure(figsize=(7,4))
sns.clustermap(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)','target_name']].corr(),annot=True,cmap='summer')
```

## Let's look at a more intersting IRIS example

```python
# more here
# https://www.python-graph-gallery.com/404-dendrogram-with-heat-map

url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
```

```python
df
```

```python
# Normalize
sns.clustermap(df, z_score=1)
plt.show()
```

```python
# plot with correlation distance
sns.clustermap(df, metric="correlation", standard_scale=1)
plt.show()

# plot with euclidean distance
sns.clustermap(df, metric="euclidean", standard_scale=1)
plt.show()
```
